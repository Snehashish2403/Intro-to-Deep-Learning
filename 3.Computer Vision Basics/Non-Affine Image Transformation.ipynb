```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline

image=cv2.imread("Speed_Sign_View_2.jpg")

cv2.imshow("Original Image",image)
cv2.waitKey()
cv2.destroyAllWindows()

```

```python
image=mpimg.imread('Speed_Sign_View_2.jpg')
plt.imshow(image)
```

```python
height,width=image.shape[:2]
height
```

```python
width
```

```python
source_points=np.float32([[200,60],[450,150],[520,500],[170,470]])
destination_points=np.float32([[0,0],[width,0],[width,height],[0,height]])

pers=cv2.getPerspectiveTransform(source_points,destination_points)
warped_img=cv2.warpPerspective(image,pers,(width,height))

cv2.imshow("Warped Image",warped_img)
cv2.waitKey()
cv2.destroyAllWindows()
```

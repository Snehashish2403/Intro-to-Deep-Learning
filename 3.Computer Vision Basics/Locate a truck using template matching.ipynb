```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2

image=cv2.imread("Find_Truck_color.jpg")

cv2.imshow("Find the truck!",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(image)
```

```python
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray,cmap='gray')
```

```python
template=cv2.imread('Truck_template.jpg',0)
cv2.imshow("Truck Template!",template)
cv2.waitKey()
cv2.destroyAllWindows()

w,h=template.shape[::-1]
```

```python
#Perform template matching
result=cv2.matchTemplate(image_gray,template,cv2.TM_CCOEFF_NORMED)
result.shape
```

```python
cv2.imshow('Matched Results!',result)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
min_val,max_val,min_loc,max_loc=cv2.minMaxLoc(result)
max_loc
```

```python
top_left=max_loc
bottom_right=(top_left[0]+w,top_left[1]+h)
cv2.rectangle(image,top_left,bottom_right,(0,0,255),5)

cv2.imshow('Truck located!',image)
cv2.waitKey()
cv2.destroyAllWindows()
```

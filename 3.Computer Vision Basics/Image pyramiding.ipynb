```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2

image=cv2.imread("truck_free_s.jpg")

cv2.imshow("Original Image",image)
cv2.waitKey()
cv2.destroyAllWindows()

plt.imshow(image)
```

```python
#Perform pyramiding
scaled_down_1=cv2.pyrDown(image)
scaled_down_1.shape
```

```python
scaled_down_2=cv2.pyrDown(scaled_down_1)
scaled_down_2.shape
```

```python
scaled_down_3=cv2.pyrDown(scaled_down_2)
scaled_down_3.shape
```

```python
cv2.imshow('Original',image)
cv2.imshow('Smaller #1',scaled_down_1)
cv2.imshow('Smaller #2',scaled_down_2)
cv2.imshow('Smaller #3',scaled_down_3)

cv2.waitKey()
cv2.destroyAllWindows()
```

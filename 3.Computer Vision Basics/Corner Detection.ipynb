```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2

image=cv2.imread("truck_free_s.jpg")

cv2.imshow("Find the truck!",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray,cmap='gray')
```

```python
corners=cv2.cornerHarris(image_gray,5,3,0.1)
plt.imshow(corners,cmap='gray')
plt.show()
```

```python
corners_dilate=cv2.dilate(corners,np.ones((8,8),np.uint8),iterations=1)
plt.imshow(corners_dilate,cmap='gray')
```

```python
#Use this dilated corners image to mark the corners in original image
image[corners_dilate>0.01*corners_dilate.max()]=[0,255,0]

cv2.imshow('Image with corners',image)
cv2.waitKey()
cv2.destroyAllWindows()
```

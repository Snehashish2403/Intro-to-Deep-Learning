```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline

image=cv2.imread("image_lane_c.jpg")

cv2.imshow("Original Image",image)
cv2.waitKey()
cv2.destroyAllWindows()

height,width=image.shape[:2]
```

```python
plt.imshow(image)
```

```python
image_gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray,cmap='gray')
```

```python
ROI=np.array([[(0,height),(400,350),(600,350),(width,height)]],dtype=np.int32)
blank=np.zeros_like(image_gray)

mask=cv2.fillPoly(blank,ROI,255)

masked_image=cv2.bitwise_and(image_gray,mask)
```

```python
plt.imshow(masked_image)
```

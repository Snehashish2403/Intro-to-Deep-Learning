```python
import matplotlib.image as mping
import matplotlib.pyplot as plt
import numpy as np
import cv2

image=cv2.imread("Self_Driving_Car.jpg")

image.shape

print("Height= ",int(image.shape[0])," pixels")
print("Width= ",int(image.shape[1])," pixels")

cv2.imshow("Self Driving Car!",image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
plt.imshow(image)
image.shape
```

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
cv2.imshow("Self Driving Car!",gray_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
hsv_image=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
cv2.imshow("Color Image",image)
cv2.imshow("HSV Image",hsv_image)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
#Hue Channel
plt.imshow(hsv_image[:,:,0])
plt.title("Hue Channel")
```

```python
#Saturation Channel
plt.imshow(hsv_image[:,:,1])
plt.title("Saturation Channel")
```

```python
#Value Channel
plt.imshow(hsv_image[:,:,2])
plt.title("Value Channel")
```

```python
#Open CV uses BGR instead of RGB
B,G,R=cv2.split(image)

B.shape
```

```python
G.shape
```

```python
R.shape
```

```python
cv2.imshow("Blue Channel",B)
cv2.waitKey()
cv2.destroyAllWindows()
#In this case we have passed a 2D matrix and thus open cv treats it as a grayscale image
```

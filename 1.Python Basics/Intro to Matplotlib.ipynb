# Draw a line graph

```python
from pylab import randn

x=randn(100)
y=randn(100)

plt.scatter(x,y)
plt.xlabel('Age')
plt.show()
```

```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np

x=np.arange(-3,3,0.01)

plt.plot(x,norm.pdf(x))
plt.show()
```

```python
plt.plot(x,norm.pdf(x))
plt.plot(x,norm.pdf(x,1.0,0.5))
plt.show()
```
